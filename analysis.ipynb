```python
print("Duplicates: ", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
# Converting IDs to string as they dont need math operations
df['Transaction_ID'] = df['Transaction_ID'].astype(str)
df['Customer_ID'] = df['Customer_ID'].astype(str)
df['Phone'] = df['Phone'].astype(str).str.replace('.0','', regex=False)

# Converting Zipcode to string
df['Zipcode'] = df['Zipcode'].astype(str).str.replace('.0','', regex=False)

# Converting Date to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Converting Year to int
df['Year'] = df['Year'].fillna(0).astype(int)

# Converting Amount & Total_Amount to numeric
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Total_Amount'] = pd.to_numeric(df['Total_Amount'], errors='coerce')
```

```python
# Standardizing gender
df['Gender'] = df['Gender'].str.strip().str.capitalize().replace({
    'M': 'Male', 'F': 'Female', 'Other': 'Other'
})

# Standardizing payment methods
df['Payment_Method'] = df['Payment_Method'].str.strip().str.title()

# Removing spaces and unifying text columns
for col in ['City','State','Country','Product_Category','Product_Brand','Product_Type']:
    df[col] = df[col].astype(str).str.strip().str.title()
```

---------UNIVARIATE ANALYSIS------------

Now, I am trying to analyse some useful variables like--

Age → Understand customer age distribution and dominant age groups.

Gender → Gender share of OmniMart’s customer base.

Income → Spread of customer income levels (helps in pricing & targeting).

Customer_Segment → Which customer segments drive business.

Total_Purchases → Frequency of purchases (loyalty vs. one-time buyers).

Product_Category → Identify best-selling product categories.

Ratings → Overall customer satisfaction trends.

Payment_Method → Popular payment options (cards, cash, PayPal, etc.).

```python

import warnings

# --- Setup and Configuration ---
warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid", palette="rocket", rc={'figure.facecolor': '#f5e9eb', 'axes.facecolor': '#f5e9eb'})
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 12



# =================================================================================
# 1. Analysis of Numerical Variables
# Goal: Understand the distribution, central tendency, and spread of key numerical metrics.
# =================================================================================
print("\n--- Analyzing Numerical Variables ---")

fig, axes = plt.subplots(2, 2, figsize=(20, 16))
fig.suptitle('Distribution of Numerical Variables', fontsize=24)

# a) Age Distribution
sns.histplot(df['Age'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Customer Age Distribution')
axes[0, 0].axvline(df['Age'].mean(), color='red', linestyle='--', label=f"Mean: {df['Age'].mean():.2f}")
axes[0, 0].axvline(df['Age'].median(), color='green', linestyle='-', label=f"Median: {df['Age'].median():.2f}")
axes[0, 0].legend()

# b) Transaction Amount Distribution
sns.histplot(df['Amount'], kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Transaction Amount Distribution')
axes[0, 1].set_xlabel('Amount per Transaction ($)')
axes[0, 1].axvline(df['Amount'].mean(), color='red', linestyle='--', label=f"Mean: {df['Amount'].mean():.2f}")
axes[0, 1].axvline(df['Amount'].median(), color='green', linestyle='-', label=f"Median: {df['Amount'].median():.2f}")
axes[0, 1].legend()

# c) Total Purchases Distribution
sns.histplot(df['Total_Purchases'], kde=True, ax=axes[1, 0], color='teal', bins=20)
axes[1, 0].set_title('Distribution of Total Purchases per Customer')
axes[1, 0].set_xlabel('Number of Purchases')
axes[1, 0].axvline(df['Total_Purchases'].mean(), color='red', linestyle='--', label=f"Mean: {df['Total_Purchases'].mean():.2f}")
axes[1, 0].axvline(df['Total_Purchases'].median(), color='green', linestyle='-', label=f"Median: {df['Total_Purchases'].median():.2f}")
axes[1, 0].legend()

# d) Customer Ratings Distribution
sns.histplot(df['Ratings'], ax=axes[1, 1], color='purple', discrete=True)
axes[1, 1].set_title('Overall Customer Satisfaction Ratings')
axes[1, 1].set_xlabel('Rating (1-5)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("univariate_analysis1.png", dpi=300, bbox_inches="tight")
plt.show()

# Print summary statistics
print("\nSummary Statistics for Numerical Variables:")
print(df[['Age', 'Amount', 'Total_Purchases', 'Ratings']].describe())

# =================================================================================
# 2. Analysis of Categorical Variables
# Goal: Understand the frequency and proportion of different categories.
# =================================================================================
print("\n--- Analyzing Categorical Variables ---")

# --- Top 10 Product Categories and Brands ---
fig, axes = plt.subplots(1, 2, figsize=(22, 10))
fig.suptitle('Top 10 Product Categories & Brands by Sales Volume', fontsize=24)

# a) Top 10 Product Categories
top_categories = df['Product_Category'].value_counts().nlargest(10)
sns.barplot(x=top_categories.index, y=top_categories.values, ax=axes[0], palette='viridis')
axes[0].set_title('Top 10 Best-Selling Product Categories')
axes[0].set_xlabel('Product Category')
axes[0].set_ylabel('Number of Transactions')
axes[0].tick_params(axis='x', rotation=45)

# b) Top 10 Product Brands
top_brands = df['Product_Brand'].value_counts().nlargest(10)
sns.barplot(x=top_brands.index, y=top_brands.values, ax=axes[1], palette='plasma')
axes[1].set_title('Top 10 Best-Selling Product Brands')
axes[1].set_xlabel('Product Brand')
axes[1].set_ylabel('Number of Transactions')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("univariate_analysis2.png", dpi=300, bbox_inches="tight")
plt.show()

# --- Customer Demographics and Preferences ---
fig, axes = plt.subplots(2, 2, figsize=(20, 18))
fig.suptitle('Customer Demographics and Preferences', fontsize=24)

# c) Gender Distribution
sns.countplot(x='Gender', data=df, ax=axes[0, 0], palette='coolwarm')
axes[0, 0].set_title('Customer Gender Distribution')

# d) Income Level Distribution
sns.countplot(x='Income', data=df, ax=axes[0, 1], palette='crest', order=['Low', 'Medium', 'High'])
axes[0, 1].set_title('Customer Income Levels')

# e) Customer Segment Distribution
sns.countplot(x='Customer_Segment', data=df, ax=axes[1, 0], palette='magma', order=['New', 'Regular', 'Premium'])
axes[1, 0].set_title('Customer Segment Distribution')

# f) Payment Method Popularity
sns.countplot(x='Payment_Method', data=df, ax=axes[1, 1], palette='cubehelix')
axes[1, 1].set_title('Popular Payment Methods')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("univariate_analysis3.png", dpi=300, bbox_inches="tight")
plt.show()

print("\n--- Univariate analysis complete. ---")

```


This initial analysis provides a clear snapshot of OmniMart's core customer and their buying habits. Our customer base is primarily young (18-35), financially moderate, and digitally native. They are highly satisfied with their purchases, giving overwhelmingly positive ratings. The business is heavily driven by the **Electronics** and **Clothing** categories, with iconic brands like Apple and Nike leading sales. Our strategic focus should be on deepening engagement with this key demographic, increasing average order value, and leveraging our high customer satisfaction as a competitive advantage.



#### **Insight 1: A Youthful and Financially Moderate Customer Base**

* **Finding:** The data clearly shows our customer age is heavily skewed towards the 18-35 range. This group predominantly falls into the 'Medium' income bracket, with an average transaction amount of around $255. This paints a picture of a customer who is trendy, tech-savvy, and values a good deal.

* **Actionable Implementations:**
    * **Targeted Marketing:** Focus marketing efforts on platforms popular with this demographic, such as Instagram, TikTok, and YouTube. Use messaging that highlights value, current trends, and modern aesthetics.
    * **Flexible Payments:** Integrate "Buy Now, Pay Later" (BNPL) services like Klarna or Afterpay at checkout to make larger purchases more accessible and reduce checkout friction.
    * **Product Curation:** Ensure the product mix is constantly updated to reflect the latest trends, especially in clothing and personal electronics.

---

#### **Insight 2: Electronics & Clothing are the Undisputed Sales Kings**

* **Finding:** 'Electronics' and 'Clothing' are by far the most purchased product categories. Within these, powerhouse brands like 'Apple', 'Samsung', 'Nike', and 'Adidas' dominate sales, indicating strong brand loyalty and specific market demand.

* **Actionable Implementations:**
    * **Inventory Management:** Prioritize high stock levels for these top-selling brands and categories to prevent stockouts and lost sales. Use this data for accurate demand forecasting.
    * **Strategic Partnerships:** Strengthen relationships with these key brands. Negotiate for co-branded marketing campaigns, exclusive product launches, or early access deals to position OmniMart as the premier destination for these items.
    * **Increase Average Order Value (AOV):** Create targeted cross-selling opportunities. For example, implement "Complete Your Setup" bundles for electronics (e.g., laptop + mouse + case) or "Shop the Look" features for clothing.

---

#### **Insight 3: Overwhelmingly Positive Customer Satisfaction**

* **Finding:** Customer `Ratings` are heavily concentrated at 4 and 5 stars. This is a significant strength, indicating that our core operations—from product quality to shipping—are successfully meeting or exceeding customer expectations.

* **Actionable Implementations:**
    * **Leverage Social Proof:** Prominently display customer ratings and testimonials on product pages, in advertisements, and in email marketing. High ratings are a powerful tool for building trust and driving conversions.
    * **Automate "Top-Rated" Features:** Create dynamic sections on the website and in the app for "Customer Favorites" or "Top-Rated This Week" to help guide new customers to products that are proven to satisfy.
    * **Isolate and Address Negative Feedback:** While rare, every 1 or 2-star rating is an opportunity. Implement a workflow for customer service to proactively reach out to these customers to resolve their issues, potentially turning a negative experience into a positive one.

---

#### **Insight 4: Digital Payments are the Norm, but Flexibility is Crucial**

* **Finding:** Credit and Debit cards are the most common payment methods, which aligns with our digitally native customer base. However, 'Cash' and 'PayPal' still account for a significant number of transactions, indicating a need for diverse payment options.

* **Actionable Implementations:**
    * **Optimize the Digital Checkout:** Ensure the credit/debit card payment process is as seamless and secure as possible, including support for saving card details for future purchases.
    * **Promote Digital Wallets:** Encourage the use of PayPal, Apple Pay, or Google Pay. These methods often increase conversion rates due to their simplicity and security.
    * **Retain Flexibility:** Continue to support a variety of payment methods. The goal is to remove any friction at checkout, as a lack of a customer's preferred option is a common reason for cart abandonment.

```python
# For correlation and scatter plots, we need to map the categorical 'Income' to a numerical scale.
# This allows us to treat it as a quantitative variable for this specific analysis.
income_mapping = {'Low': 0, 'Medium': 1, 'High': 2}
df['Income_Num'] = df['Income'].map(income_mapping)
```

```python
# Correlation Matrix Overview --
# Insight: A quick heatmap to quantify the linear relationships between our variables of interest.
numerical_cols = ['Age', 'Total_Amount', 'Total_Purchases', 'Ratings']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=.5,
            xticklabels=['Age', 'Income', 'Total Amount', 'Total Purchases', 'Ratings'],
            yticklabels=['Age', 'Income', 'Total Amount', 'Total Purchases', 'Ratings'])
plt.title('Correlation Matrix of Key Numerical Variables')


plt.savefig("correlation_matrix.png", dpi=300, bbox_inches="tight", facecolor="#f5e9eb")
plt.show()
print("Generated: Correlation Heatmap")
```

##Insights-
Total Purchases and Total Amount show a strong positive correlation (0.65), meaning customers who make more purchases tend to spend more overall.

Age and Ratings have a mild positive correlation (0.17), suggesting older customers may give slightly higher ratings.

Income shows very weak or negligible correlation with other variables, implying spending and satisfaction are not strongly tied to reported income levels.

Here, I am trying to find any relation between amount spend and age of the customer. This will help us to identify our customers of a certain age group who has a potential to buy more than others.

##Insigts-
Spending is highly variable across all ages, with no strong linear trend between age and total spending.

Younger customers (around 18–30) show both very high and very low spending, indicating diverse purchasing behaviors in this group.

Middle-aged and older customers (40+) generally cluster around lower to mid spending levels, with fewer extreme outliers compared to younger groups.

Since, I didnt got any proper relation between spending and age group. I am now trying to find a relation between spending ability and other factors to find our most loyal customers. Once the loyal customers are found, we can launch bonus discounts and gift/prize for them, urging low transcation value customers to buy more, hence increasing our loyal customer base.

```python

df_sample = df.sample(n=1000, random_state=42)


fig, axes = plt.subplots(3, 2, figsize=(15, 15), facecolor="#FFF5FF")
fig.suptitle('Analysis of Numerical Relationships', fontsize=24, y=1.02)

# Plot 1: Age vs. Total_Amount
sns.regplot(ax=axes[0, 0], data=df_sample, x='Age', y='Total_Amount',
            line_kws={"color": "red"}, scatter_kws={'alpha':0.5})
axes[0, 0].set_title('Spending Behavior Across Different Ages')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Total Amount ($)')

# Plot 2: Income vs. Total_Amount
sns.regplot(ax=axes[0, 1], data=df_sample, x='Income_Num', y='Total_Amount',
            line_kws={"color": "red"}, scatter_kws={'alpha':0.5})
axes[0, 1].set_title('Spending by Income Level')
axes[0, 1].set_xlabel('Income Level')
axes[0, 1].set_ylabel('Total Amount ($)')
axes[0, 1].set_xticks(ticks=list(income_mapping.values()), labels=list(income_mapping.keys()))

# Plot 3: Total_Purchases vs. Total_Amount
sns.regplot(ax=axes[1, 0], data=df_sample, x='Total_Purchases', y='Total_Amount',
            line_kws={"color": "red"}, scatter_kws={'alpha':0.5})
axes[1, 0].set_title('Purchase Frequency vs. Total Spending')
axes[1, 0].set_xlabel('Number of Purchases')
axes[1, 0].set_ylabel('Total Amount ($)')

# Plot 4: Ratings vs. Total_Amount
sns.regplot(ax=axes[1, 1], data=df_sample, x='Ratings', y='Total_Amount',
            line_kws={"color": "red"}, scatter_kws={'alpha':0.5}, x_jitter=.2) # Jitter for discrete x-axis
axes[1, 1].set_title('Do Higher Spenders Give Better Ratings?')
axes[1, 1].set_xlabel('Customer Rating (1-5)')
axes[1, 1].set_ylabel('Total Amount ($)')

# Plot 5: Age vs. Income
sns.regplot(ax=axes[2, 0], data=df_sample, x='Age', y='Income_Num',
            line_kws={"color": "red"}, scatter_kws={'alpha':0.5}, y_jitter=.2) # Jitter for discrete y-axis
axes[2, 0].set_title('Income Variation by Age Group')
axes[2, 0].set_xlabel('Age')
axes[2, 0].set_ylabel('Income Level')
axes[2, 0].set_yticks(ticks=list(income_mapping.values()), labels=list(income_mapping.keys()))

# Plot 6: Ratings vs. Age
sns.regplot(ax=axes[2, 1], data=df_sample, x='Age', y='Ratings',
            line_kws={"color": "red"}, scatter_kws={'alpha':0.5}, y_jitter=.2) # Jitter for discrete y-axis
axes[2, 1].set_title('Rating Behavior Across Age Groups')
axes[2, 1].set_xlabel('Age')
axes[2, 1].set_ylabel('Customer Rating (1-5)')

plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
plt.savefig("bivariate_num_vs_num.png", dpi=300, bbox_inches="tight", facecolor="#FFF5FF")

print("Generated: 3x2 grid of regression plots.")
```

##Insights-
1. Age vs. Spending → Spending is spread widely across ages with no strong trend, but younger and middle-aged groups show more high-value purchases.

2. Income vs. Spending → Income level has little effect on spending, as low, medium, and high income groups show similar total amounts.

3. Purchases vs. Spending → A strong positive relationship exists: customers with more purchases clearly spend more overall.

4. Spending vs. Ratings → Ratings are not strongly tied to spending, as high and low spenders both give mixed ratings.

5. Age vs. Income → Younger customers tend to report higher income, with a slight decline in income levels for older age groups.

6. Age vs. Ratings → Older customers give slightly higher ratings on average, but overall variation is high across all age groups.

```python
import numpy as np
# using a sample to make plots render faster and less cluttered
df_sample2 = df.sample(n=1000, random_state=42)

fig, axes = plt.subplots(2, 2, figsize=(18, 15), facecolor="#FFF5FF")
fig.suptitle('Numerical vs Categorical analysis', fontsize=24, y=1.02)

# plot 1: Age vs. Customer_Segment → boxplot/violin plot (are certain segments older/younger?).
sns.violinplot(
    ax=axes[0, 0],
    data=df_sample2,
    x='Age',
    y='Customer_Segment',
    inner="box",
    scale="width",
    palette="Set2",
    linewidth=1)
axes[0, 0].set_title('Customer Segmentation Across Different Ages')
axes[0, 0].set_xlabel('Age')
axes[0, 0].set_ylabel('Segment')

# plot 2: Total_Amount vs. Product_Category → boxplot/barplot (which categories drive higher spending?).
sns.boxplot(
    ax=axes[0,1],
    data=df_sample2,
    x="Product_Category",
    y="Total_Amount",
    order=None,
    palette="Set2",
    orient="v")
axes[0, 1].set_title('Spending Distribution by Product Category')
axes[0, 1].set_xlabel('Category of Product')
axes[0, 1].set_ylabel('Total Amount')

# plot 3: Amount vs. Gender → boxplot (are there differences in spending between genders?).
sns.barplot(
    ax=axes[1,0],
    data=df_sample2,
    x="Gender",
    y="Amount",
    estimator=np.mean,
    ci="sd",
    palette="Set2",
    capsize=0.2,
    errwidth=1.5,
    dodge=True
)

axes[1, 0].set_title('Average Spending by Gender')
axes[1, 0].set_ylabel('Amount')
axes[1, 0].set_xlabel('Gender')

fig.delaxes(axes[1, 1])
plt.savefig("bivariate_num_vs_cat.png", dpi=300, bbox_inches="tight", facecolor="#f5e9eb")

```

##Insights-
1. Customer segmentation by age shows that younger customers dominate the "New" and "Regular" segments, while "Premium" customers are more evenly spread across middle and older ages.

2. Spending distribution by product category indicates relatively similar spending levels across categories, with Electronics and Books showing slightly higher upper outliers.

3. Average spending by gender is nearly the same for males and females, with only marginal differences and overlapping variability.

```python
# using a sample to make plots render faster and less cluttered
df_sample3 = df.sample(n=1000, random_state=42)

fig, axes = plt.subplots(2, 2, figsize=(23, 23), facecolor="#f5e9eb")
fig.suptitle('Categorical vs Categorical analysis', fontsize=24, y=1.02)

# plot 1: Customer_Segment vs. Payment_Method → heatmap/crosstab (which segment prefers which payment method?).
ct1 = pd.crosstab(df_sample3["Customer_Segment"], df_sample3["Payment_Method"], normalize="index") * 100
sns.heatmap(ct1, annot=True, fmt=".1f", cmap="YlGnBu", cbar_kws={'label': 'Percentage (%)'}, ax=axes[0,0])
axes[0,0].set_title("Customer Segment vs Payment Method (%)")
axes[0,0].set_xlabel("Payment Method")
axes[0,0].set_ylabel("Customer Segment")

# plot 2: Product_Category vs. Shipping_Method → stacked bar chart (which shipping options dominate per category?).
df_sample3_clean = df_sample3.dropna(subset=["Product_Category", "Shipping_Method"])
ct2 = pd.crosstab(df_sample3_clean["Product_Category"], df_sample3["Shipping_Method"], normalize="index") * 100
ct2.plot(kind="bar", stacked=True, colormap="Set2", ax=axes[0,1])
axes[0,1].set_title("Product Category vs Shipping Method (Stacked %)")
axes[0,1].set_xlabel("Product Category")
axes[0,1].set_ylabel("Percentage (%)")
axes[0,1].tick_params(axis='x', rotation=45)
axes[0,1].legend(title="Shipping Method", bbox_to_anchor=(1.05, 1), loc='upper left')

# plot 3: Order_Status vs. Payment_Method → heatmap (are certain payment methods linked to more cancellations/refunds?).
ct3 = pd.crosstab(df_sample3["Order_Status"], df_sample3["Payment_Method"], normalize="index") * 100
sns.heatmap(ct3, annot=True, fmt=".1f", cmap="OrRd", cbar_kws={'label': 'Percentage (%)'}, ax=axes[1,0])
axes[1,0].set_title("Order Status vs Payment Method (%)")
axes[1,0].set_xlabel("Payment Method")
axes[1,0].set_ylabel("Order Status")

# plot 4: Gender vs. Feedback (Positive/Negative/Neutral) → grouped barplot (do feedback patterns differ by gender?).
ct4 = pd.crosstab(df_sample3["Gender"], df_sample3["Feedback"], normalize="index") * 100
ct4_reset = ct4.reset_index().melt(id_vars="Gender", var_name="Feedback", value_name="Percentage")
sns.barplot(data=ct4_reset, x="Gender", y="Percentage", hue="Feedback", palette="Accent", ax=axes[1,1])
axes[1,1].set_title("Gender vs Feedback (%)")
axes[1,1].set_xlabel("Gender")
axes[1,1].set_ylabel("Percentage (%)")
axes[1,1].legend(title="Feedback", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.savefig("bivariate_cat_vs_cat.png", dpi=300, bbox_inches="tight", facecolor="#fccccc")

```

```python
# Using sample for faster plotting
df_sample4 = df.sample(n=1000, random_state=42)



# Defining the key columns used across the 6 plots
cols_needed = [
    "Age", "Income", "Total_Amount",
    "Customer_Segment", "Payment_Method",
    "Product_Category", "Shipping_Method", "Order_Status",
    "Gender", "Feedback", "Ratings",
    "Month", "Total_Purchases"
]

# Dropping rows with NaN in ANY of these columns
df_sample4 = df_sample4.dropna(subset=cols_needed)

# Setting up subplots
fig, axes = plt.subplots(3, 2, figsize=(25,30))

# plot 1: Age x Income x Total Amount --
sns.scatterplot(
    ax=axes[0,0],
    data=df_sample4,
    x="Age",
    y="Total_Amount",
    hue="Income",
    alpha=0.7
)
axes[0,0].set_title("Spending by Age & Income")
axes[0,0].set_xlabel("Age")
axes[0,0].set_ylabel("Total Amount")

# plot 2: Gender x Feedback x Ratings
sns.boxplot(
    ax=axes[0, 1],
    data=df_sample4,
    x="Feedback",
    y="Ratings",
    hue="Gender",
    palette="Accent"
)
axes[0, 1].set_title("Ratings by Feedback & Gender")
axes[0, 1].set_xlabel("Feedback")
axes[0, 1].set_ylabel("Ratings")

# plot 3: Customer Segments vs Payment methods vs total amount
sns.barplot(
    ax=axes[1, 0],
    data=df_sample4,
    x="Customer_Segment",
    y="Total_Amount",
    hue="Payment_Method",
    estimator=np.mean,
    ci="sd",
    capsize=0.2,
    errwidth=1.2,
    dodge=True,
    palette="Set2"
)
axes[1, 0].set_title("Average Spending by Segment & Payment Method")
axes[1, 0].set_xlabel("Customer Segment")
axes[1, 0].set_ylabel("Avg Total Amount")

# plot 4: Month x product category x total amount
sns.lineplot(
    ax=axes[1, 1],
    data=df_sample4,
    x="Month",
    y="Total_Amount",
    hue="Product_Category",
    estimator="sum",
    errorbar=None,
)
axes[1, 1].set_title("Monthly Sales Trends by Product Category")
axes[1, 1].set_xlabel("Month")
axes[1, 1].set_ylabel("Total Amount")
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right');

# plot 5: Product category x shipping methods x order status
sns.countplot(
    ax=axes[2,0],
    data=df_sample4,
    x="Product_Category",
    hue="Shipping_Method"
)
axes[2,0].set_title("Shipping Methods by Product Category")
axes[2,0].set_xlabel("Product Category")
axes[2,0].set_ylabel("Order Count")
axes[2,0].tick_params(axis='x', rotation=45)

# plot 6: Customer segments x age x total purchase
sns.scatterplot(
    ax=axes[2,1],
    data=df_sample4,
    x="Age",
    y="Total_Purchases",
    hue="Customer_Segment",
    size="Total_Purchases",
    alpha=0.7,
    sizes=(20,200)
)
axes[2,1].set_title("Purchase Frequency by Age & Segment")
axes[2,1].set_xlabel("Age")
axes[2,1].set_ylabel("Total Purchases")

plt.savefig("multivariate.png", dpi=300, bbox_inches="tight", facecolor="#f5e9eb")



```

```python
# --- 2. Identify Outliers using IQR ---
# Calculate Q1, Q3, and IQR for the 'Popularity' column
Q1 = product_popularity['Popularity'].quantile(0.25)
Q3 = product_popularity['Popularity'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"--- IQR Outlier Analysis ---")
print(f"First Quartile (Q1): {Q1}")
print(f"Third Quartile (Q3): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Outlier Upper Bound (Anything > {upper_bound:.2f} is an outlier)")
print("\n")

outlier_products = product_popularity[product_popularity['Popularity'] > upper_bound]

print("--- Identified Outlier Products (Hyper-Popular) ---")
if outlier_products.empty:
    print("No outliers found based on the IQR method.")
else:
    print(outlier_products)
print("\n")
```

Now , I am trying to find out our worst products,i.e, products that have above average transcation value, but have a low rating. These products are the ones that cause harm to our customer base. Will try to identify and improve these products.

```python


# --- 1. Calculate Product-Level Statistics ---
product_stats = df.groupby('products').agg(
    Average_Rating=('Ratings', 'mean'),
    Average_Amount=('Amount', 'mean'),
    Purchase_Count=('Transaction_ID', 'count')
).reset_index()

# --- 2. Calculate Global Averages ---
global_avg_rating = product_stats['Average_Rating'].mean()
global_avg_amount = product_stats['Average_Amount'].mean()

# --- 3. Identify Underperforming Products ---
# Criteria: Below-average rating AND above-average amount
underperforming_df = product_stats[
    (product_stats['Average_Rating'] < global_avg_rating) &
    (product_stats['Average_Amount'] > global_avg_amount)
].copy()

# --- 4. Create a "Problem Score" to Rank the Worst ---
# We normalize the metrics to create a fair score.
# Higher score = worse product
underperforming_df['Rating_Deviation'] = (global_avg_rating - underperforming_df['Average_Rating']) / global_avg_rating
underperforming_df['Amount_Deviation'] = (underperforming_df['Average_Amount'] - global_avg_amount) / global_avg_amount
underperforming_df['Problem_Score'] = underperforming_df['Rating_Deviation'] + underperforming_df['Amount_Deviation']

# Get the top 10 worst products
top_10_worst_df = underperforming_df.sort_values(by='Problem_Score', ascending=False).head(10)

print("--- Top 10 Worst-Performing Products (High Price, Low Rating) ---")
print(top_10_worst_df[['products', 'Average_Rating', 'Average_Amount', 'Problem_Score']])

# --- 5. Create the Clean Bar Chart ---
print("\nGenerating bar chart for the worst products...")
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(16, 10))
fig.set_facecolor('white')

# Create a color palette from yellow to red
palette = sns.color_palette("YlOrRd_r", n_colors=len(top_10_worst_df))

# Plot the data
barplot = sns.barplot(
    x='Average_Rating',
    y='products',
    data=top_10_worst_df.sort_values(by='Average_Rating', ascending=True),
    palette=palette,
    orient='h'
)

# --- 6. Add Data Labels and Polish the Aesthetics ---
# Add labels showing both rating and average amount
for i, (index, row) in enumerate(top_10_worst_df.sort_values(by='Average_Rating', ascending=True).iterrows()):
    rating_text = f' {row["Average_Rating"]:.2f} Rating'
    amount_text = f'(${row["Average_Amount"]:.2f} Avg. Price)'
    ax.text(row['Average_Rating'], i, rating_text, color='black', ha="left", va="center", fontsize=12, weight='bold')
    ax.text(0.05, i, amount_text, color='black', ha="left", va="center", fontsize=11)

# Set a clear title and subtitle
ax.set_title("OmniMart's Top 10 Most Problematic Products", fontsize=20, weight='bold', pad=20)
plt.suptitle("These products have high average prices but fail to meet customer expectations.",
             fontsize=14, y=0.92, color='dimgray')

# Customize labels and limits
ax.set_xlabel('Average Customer Rating (The Lower, The Worse)', fontsize=14, labelpad=10)
ax.set_ylabel('Product', fontsize=14, labelpad=10)
ax.set_xlim(0, 5) # Ratings are on a 1-5 scale

# Visual reference for the company-wide average rating
ax.axvline(x=global_avg_rating, color='crimson', linestyle='--', linewidth=2, label=f'Avg. Product Rating ({global_avg_rating:.2f})')
ax.legend()

# Remove unnecessary visual clutter
sns.despine(left=True, bottom=True)
ax.grid(axis='x', linestyle='--', alpha=0.7)
ax.tick_params(axis='y', labelsize=12)

plt.tight_layout(rect=[0, 0, 1, 0.9])

# Save the figure
fig.savefig('top_10_worst_products.png', dpi=150, facecolor="#f5e9eb")
print("Plot successfully saved as 'top_10_worst_products.png'")

plt.show()
```
